# Q1 Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Q2 10 rows of dataset

```python
ab = pd.read_csv('AB_NYC_2019.csv')
ab.head(10)
```

# Q3 Boxplot and histogram of price column, with bins=30

```python
plt.figure(figsize=(8,3))
plt.boxplot(ab['price'], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title('Box plot of Price column')
plt.xlabel('Price')
plt.show()
```

```python
sns.histplot(data=ab, x=ab['price'], bins=30)
```

# Q4 Important libraries for z score  

```python
from scipy import stats
```

# Q5 zscore and abs value of price column

```python
price_zscore = stats.zscore(ab['price'])
abs_z = np.abs(price_zscore)
abs_z
```

# Q6 zscore is greater than 3

```python
threshold = 3
outliers = ab[abs_z>threshold]
outliers['id'].size
```

### I can't believe we have 388 data with zscore higher than 3, how does that make sense? how can there be so many data after 3 std deviations?

```python
outliers
```

# Q7 Number of outliers

```python
outliers['id'].size
```

# Q8 New dataset with removed outliers

```python
new_ab = ab[abs_z<=threshold]
new_ab
```

```python
print(ab['id'].size)
print(new_ab['id'].size)
```

```python
sns.boxplot(x=new_ab['price'])
```

# IQR method
# Q9-Q10 Quartiles

```python
desc = new_ab['price'].describe()
Q1 = desc['25%']
Q2 = desc['50%']
Q3 = desc['75%']
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print("Q1: ", Q1)
print("Q3: ", Q3)
print("IQR: ", IQR)
print("Lower bound: ", lower_bound)
print("Upper bound: ", upper_bound)
```

# Q11 Values below the lower limit

### Lower limit

```python
ab_lower_limit = new_ab[(new_ab['price'] < lower_bound)]
ab_lower_limit
```

```python
ab_lower_limit.size
```

### It makes sense, why do we see 0, because lower bound is below 0

### Upper limit

```python
ab_upper_limit = new_ab[(new_ab['price'] > upper_bound)]
ab_upper_limit.head()
```

```python
ab_upper_limit.size
```

# Q12 Dataset with all the outliers

## Q12-1 All the outliers with a single line

```python
new_ab_outliers = new_ab[(new_ab['price'] < lower_bound) | (new_ab['price'] > upper_bound)]
new_ab_outliers.head(10)
```

```python
new_ab_outliers['id'].size
```

# Q13 Clean data without outliers

```python
new_ab_no_outliers = new_ab[(new_ab['price'] >= lower_bound) & (new_ab['price'] < upper_bound)]
```

```python
new_ab_no_outliers.head(10)
```

```python
print(new_ab_no_outliers['id'].size)
print(new_ab_outliers['id'].size)
```

```python
sns.boxplot(x=new_ab_no_outliers['price'])
```

# Q13-2 Clean dataset without outliers using the complement operator

```python
new_ab_clean = new_ab[~((new_ab['price'] < lower_bound) | (new_ab['price'] > upper_bound))]
new_ab_clean.head(10)
```

```python
print(new_ab_clean['id'].size)
print(new_ab_no_outliers['id'].size)
print(new_ab['id'].size)
```

```python
ab_clean = ab[~((ab['price'] < lower_bound) | (ab['price'] > upper_bound))]
ab_clean.head(5)
```
