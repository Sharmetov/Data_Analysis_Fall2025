# 10 - Bayes' Theorem and Total Probability

### Now, that I figured out the correct conditional probability, we can easily find the "reverse" condition. Just stick to this formula
### P(R<500 IN JUNE| FLOOD) = P(FLOOD|R<500 IN JUNE) * P(R<500 IN JUNE) / P(FLOOD)

```python
P_r_500less_june_given_fl = P_fl_given_r_500less_june * P_r_500less_june / P_flood
P_r_500less_june_given_fl
```

### Total probability

$P(R<500) = P(R<500|F) * P(F) + P(R<500|F^c) * P(F^c) $

$P(R<500|F^c) = P(R<500  \cap F^c) / P(F^c) $

```python
cnt_no_flood = (kerala_floods['FLOODS'] == 0).sum()
cnt_no_flood
```

```python
P_no_flood = cnt_no_flood / N
P_no_flood 
```

```python
cnt_r_500less_june_no_fl = ((kerala_floods['JUNE_MORE_THAN_500'] == 0) & (kerala_floods['FLOODS'] == 0)).sum()
cnt_r_500less_june_no_fl
```

```python
P_r_500less_june_no_fl = cnt_r_500less_june_no_fl / N
P_r_500less_june_no_fl
```

```python
P_r_500less_june_given_no_fl = P_r_500less_june_no_fl / P_no_flood
P_r_500less_june_given_no_fl
```

$P(R<500) = P(R<500|F) * P(F) + P(R<500|F^c) * P(F^c) $

```python
P_r_500less = P_r_500less_june_given_fl * P_flood + P_r_500less_june_given_no_fl * P_no_flood
P_r_500less
```
