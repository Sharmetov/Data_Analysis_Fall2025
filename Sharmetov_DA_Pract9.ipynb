# 1 - New column JUNE_MORE_THAN_500 

```python
kerala_floods['JUNE_MORE_THAN_500'] = (kerala_floods['JUN'] > 500).astype(int)
```

```python
kerala_floods['JUNE_MORE_THAN_500'].value_counts()
```

# 2 - Updates to the column 'FLOODS'

```python
kerala_floods['FLOODS'].value_counts()
```

```python
kerala_floods['FLOODS'] = (kerala_floods['FLOODS'] == 'YES').astype(int)
```

```python
kerala_floods['FLOODS'].value_counts()
```

# 3 - Select needed columns

```python
kerala_floods[['JUNE_MORE_THAN_500', 'FLOODS']]
```

# 3 - Crosstab between these two columns: FLOODS, JUNE_MORE_THA_500

```python
crosstab = pd.crosstab(kerala_floods['JUNE_MORE_THAN_500'], kerala_floods['FLOODS'])
```
