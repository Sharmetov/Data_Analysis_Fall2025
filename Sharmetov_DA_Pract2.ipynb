#### Sample variance for Nur

```python
data['Nur'].var()
```

#### Sample variance for Nur using formula that we are familiar with

```python
X = data['Nur']
X_mean = X.mean()
Nur_var = ((X - X_mean)**2).sum() / (len(X)-1)
```

```python
Nur_var
```

#### Same results

#### Population variance for Nur

```python
X.var(ddof=0)
```

#### Population variance for Nur using formula

```python
Nur_var_ppl = ((X - X_mean) ** 2 ).sum() / len(X)
```

```python
Nur_var_ppl
```
