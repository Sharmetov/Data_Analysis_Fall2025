```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
country = ['Kazakhstan', 'Uzbekistan', 'Kyrgyzstan', 'Tajikistan']
pop_million = [20, 31, 6, 9]
colors = ['#FA8F0A','#F0DA32','#61F527','#18DE9D','#3594FF']
```

### Pie chart

```python
max_index = np.argmax(pop_million)
```

```python
explode = np.zeros(len(pop_million))
explode[max_index] = 0.1
```

```python
plt.pie(pop_million, labels=country, colors=colors, autopct = "%1.1f%%", explode=explode)
plt.title("Population of Some Countries in Central Asia")
plt.show()
```

### Histogram

```python
X = np.random.normal(150, 10, 200)
```

```python
plt.hist(X, bins=20, edgecolor = 'y', label = 'Histogram of X')
plt.title('Histogram of X')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(X, bins=20, edgecolor = 'y', label = 'Histogram of X', orientation='horizontal')
plt.title('Histogram of X')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.show()
```

### Line Plot

```python
x = np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x))
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x), label='sin(x)', color = 'y')
plt.plot(x, np.cos(x), label='cos(x)', color = 'b')
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.show()
```
